```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
```

```python
df_train = pd.read_csv('./forest-cover-type-prediction/train.csv')
```

```python
df_train.head()
```

# Data Prep

```python
df_train['Wilderness_Area'] = df_train.iloc[:,11:15].idxmax(axis=1).str.replace('Wilderness_Area','')
df_train['Soil_Type'] = df_train.iloc[:,16:55].idxmax(axis=1).str.replace('Soil_Type','')
```

```python
df_train.head()
```

```python
df_train_simple = df_train.iloc[:,:11]
extra_cols = ['Wilderness_Area', 'Soil_Type', 'Cover_Type']
df_train_simple[extra_cols] = df_train[extra_cols]
df_train_simple
```

```python
X_train = df_train_simple.iloc[:,1:12]
Y_train = df_train_simple.iloc[:,-1:]
```

# Decision Tree

```python
dtree = DecisionTreeClassifier(max_depth=4, random_state=0)
dtree.fit(X_train, Y_train)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (8,8), dpi=600)

fn=X_train.columns
cn=np.unique(Y_train.values).astype(str)
tree.plot_tree(dtree,
               feature_names = fn, 
               class_names=cn,
               filled = True)
```

```python
tbl = '\n'.join(f'{x:<40}{round(y,3)}' for x,y in zip(X_train.columns, dtree.feature_importances_))
print(tbl)
```

# Random Forest

```python
rfmod = RandomForestClassifier(max_depth=4)
Y_train_xform = np.reshape(Y_train.values, (len(Y_train)))
rfmod.fit(X_train, Y_train_xform)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 3,figsize = (5,2), dpi=3000)
for index in range(0, 3):
    tree.plot_tree(rfmod.estimators_[index],
                   feature_names = fn, 
                   class_names=cn,
                   filled = True,
                   ax = axes[index]);
    
    axes[index].set_title('Model: ' + str(index), fontsize = 11)
```
